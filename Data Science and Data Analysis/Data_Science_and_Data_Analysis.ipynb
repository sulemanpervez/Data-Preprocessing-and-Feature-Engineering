```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Student_ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [21, 22, np.nan, 24, 23],
    'Grade': [85, 90, 78, 92, np.nan],
    'Gender': ['F', 'M', 'M', 'M', 'F']
}
df = pd.DataFrame(data)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Grade'].fillna(df['Grade'].median(), inplace=True)
summary_stats = df.describe()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Grade'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation = df['Age'].corr(df['Grade'])
gender_avg_grade = df.groupby('Gender')['Grade'].mean()
```

```python
print("Data Cleaning:")
print(df)
print("\nExploration - Summary Statistics:")
print(summary_stats)
print("\nStatistical Analysis - Correlation between Age and Grade:")
print(f"Correlation: {correlation}")
print("\nStatistical Analysis - Gender-wise Average Grade:")
print(gender_avg_grade)
```

## Problem Statement

We have a dataset containing the following columns:

- **Customer_ID**: Unique identifier for each customer.
- **Age**: Age of the customer.
- **Gender**: Gender of the customer.
- **Income**: Income of the customer.
- **Product_Category**: Category of the purchased product.
- **Purchase_Amount**: Amount spent by the customer on the purchase.

Our goal is to perform data analysis to answer the following questions:

1. **What is the distribution of purchase amounts?**
2. **How do purchase amounts vary across different product categories?**
3. **Is there any correlation between customer age/income and purchase amount?**
4. **What is the average purchase amount for each gender?**
5. **Are there any insights or patterns in customer purchasing behavior that can be leveraged for targeted marketing?**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Purchase_Amount': np.random.normal(loc=100, scale=30, size=1000),
    'Product_Category': np.random.choice(['A', 'B', 'C'], size=1000),
    'Age': np.random.randint(18, 70, size=1000),
    'Income': np.random.normal(loc=50000, scale=15000, size=1000),
    'Gender': np.random.choice(['Male', 'Female'], size=1000)
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Purchase_Amount'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Purchase_Amount', by='Product_Category', figsize=(10, 6))
plt.title('Purchase Amounts Across Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_age_purchase = df['Age'].corr(df['Purchase_Amount'])
correlation_income_purchase = df['Income'].corr(df['Purchase_Amount'])
gender_avg_purchase = df.groupby('Gender')['Purchase_Amount'].mean()
```

```python
print("Data Exploration - Summary Statistics:")
print(df.describe())
print("\nStatistical Analysis - Correlation between Age/Income and Purchase Amount:")
print(f"Correlation between Age and Purchase Amount: {correlation_age_purchase}")
print(f"Correlation between Income and Purchase Amount: {correlation_income_purchase}")
print("\nStatistical Analysis - Average Purchase Amount for Each Gender:")
print(gender_avg_purchase)
```
